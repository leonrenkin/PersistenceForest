# Landscapes for 3d point clouds

Landscapes for point clouds in dimension $n$ require a function which maps every $n-1$ cycles (stored as a SignedChain object containg list of $d-1$-simplices and their orientation) onto a real number.

So far, signed_chain_volume is the only vectorisation of cycles which is implemented in arbitrary dimension. This is the volumen of the cycle itself, and NOT the interion. For a 3d point cloud, this gives the boundary area and NOT the contained volume.

```python
from persforest.cycle_rep_vectorisations import signed_chain_volume

forest.compute_generalized_landscape_family(
    cycle_func=signed_chain_volume,
    label="boundary_area")

ax = forest.plot_landscape_family(label="boundary_area")
plt.show()
```

# Evaluate landscape and barcode functionals on grid

```python
grid = np.linspace(0,3,20) #this is grid on which we want to evaluate the landscape

landscape = forest.compute_generalized_landscape_family(    
    cycle_func=signed_chain_volume,
    label="boundary_area", 
    cache_functionals=True)


#alternatively, one can load a previously computed landscape by
#landscape = forest.landscape_families[label]

landscape_vec = landscape.evaluate_on_grid(grid, levels=3)

print(f"Shape {landscape_vec.shape}")
print(landscape_vec)
```

```python
# Since we set cache_functional = True in cell above, we can load, plot and vectorize the measurement on the intervals directly

barcode_functionals = forest.barcode_functionals["boundary_area"]

max_bar = forest.max_bar()

ax = barcode_functionals[max_bar].plot(x_range = (0,2.5), y_range = (0,35))
plt.show()

barcode_functional_vec = barcode_functionals.evaluate_on_grid(bars = [max_bar], grid = grid)

print(barcode_functional_vec)
```
